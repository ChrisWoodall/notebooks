```python
import pandas as pd
import numpy as np
from scipy.stats import zscore
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Reading CSV files

df = pd.read_csv("~/Desktop/client_1567_1_workticketlength.csv", index_col='Unnamed: 0')
df
```

```python
#  averaging and filtering by zscore

# lets look at the stats first

df.head()
# print (df.describe())
# print (df['duration'].sum() / 35)
# print (df['workticket_duration'].sum() /35)

S= pd.Series(df['duration'])
fig1 = plt.plot(S)
R = (S[~((S-S.mean()).abs()>3*S.std())])

fig2 = plt.plot(R)
T = pd.Series(df['workticket_duration'])
fig3 = plt.plot(S)
Q = (T[~((T-T.mean()).abs()>3*T.std())])
fig4 = plt.plot(Q)


R.describe(percentiles=[0.95])
```
