```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
days, impressions = np.loadtxt('page-impressions.csv', unpack=True, converters=
                              {0: mdates.strpdate2num('%Y-%m-%d')})
```

```python
days[0:2]
```

```python
plt.plot_date(x=days, y=impressions, fmt="r-")
plt.show()
```
